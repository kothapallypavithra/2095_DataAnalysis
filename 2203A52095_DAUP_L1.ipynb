Plot between smoker and non smokers (20-40) age group

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/insurance.csv")

# Convert the 'smoker' column to numeric for simplicity
data['smoker'] = data['smoker'].map({'yes': 1, 'no': 0})

# Filter the data where age is between 20 and 40
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]

# Separate smokers and non-smokers
smokers = filtered_data[filtered_data['smoker'] == 1]
non_smokers = filtered_data[filtered_data['smoker'] == 0]

# Plot the data
plt.scatter(non_smokers['age'], non_smokers['expenses'], c='blue', label='Non-Smokers', alpha=0.6)
plt.scatter(smokers['age'], smokers['expenses'], c='red', label='Smokers', alpha=0.7)

# Add title, axis labels, and legend
plt.title("Scatter Plot: Smokers (20-40 Age Group)", fontsize=16,color='red')
plt.xlabel("Age", fontsize=12)
plt.ylabel("expenses ($)", fontsize=12)
plt.legend()

# Count smokers in the age group
num_smokers = smokers.shape[0]
print(f"Number of smokers aged 20-40: {num_smokers}")

# Annotate the number of smokers in the plot
plt.text(30, max(filtered_data['expenses']) - 5000, f"Smokers: {num_smokers}", fontsize=12, color='red')

# Show grid for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

3. Filter the data based on a condition and update the plot to reflect the filtered data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("/content/insurance.csv")

# Convert the 'smoker' column to numeric for simplicity
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 1, 'no': 0})

# Apply new condition: filter by charges greater than $10,000
filtered_data = insurance_data[insurance_data['expenses'] > 10000]

# Separate smokers and non-smokers within the filtered data
smokers = filtered_data[filtered_data['smoker'] == 1]
non_smokers = filtered_data[filtered_data['smoker'] == 0]

# Create the scatter plot for the filtered data
plt.figure(figsize=(8, 6))

# Plot for non-smokers in blue
plt.scatter(non_smokers['age'], non_smokers['expenses'], c='blue', marker='o', label='Non-Smokers', alpha=0.6)

# Plot for smokers in red
plt.scatter(smokers['age'], smokers['expenses'], c='red', marker='+', label='Smokers', alpha=0.7)

# Add title, axis labels, and legend
plt.title("Insurance Charges > $10,000: Smokers and Non-Smokers", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Insurance Charges ($)", fontsize=12)
plt.legend()

# Display grid
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

plot for smokers and non smokers above 39 and below 39

```python
import matplotlib.pyplot as plt

# Convert smoker column to categorical for clarity
data['smoker'] = data['smoker'].map({1: 'yes', 0: 'no'})

# Split the data based on conditions
above_39_smokers = data[(data['smoker'] == 'yes') & (data['age'] > 39)]
below_39_smokers = data[(data['smoker'] == 'yes') & (data['age'] <= 39)]
above_39_non_smokers = data[(data['smoker'] == 'no') & (data['age'] > 39)]
below_39_non_smokers = data[(data['smoker'] == 'no') & (data['age'] <= 39)]

# Generate a scatter plot
plt.figure(figsize=(12, 8))

# Plot smokers above 39
plt.scatter(above_39_smokers['age'], above_39_smokers['expenses'], color='red', label='Smokers (Age > 39)', alpha=0.7)

# Plot smokers below 39
plt.scatter(below_39_smokers['age'], below_39_smokers['expenses'], color='blue', label='Smokers (Age <= 39)', alpha=0.7)

# Plot non-smokers above 39
plt.scatter(above_39_non_smokers['age'], above_39_non_smokers['expenses'], color='green', label='Non-Smokers (Age > 39)', alpha=0.7)

# Plot non-smokers below 39
plt.scatter(below_39_non_smokers['age'], below_39_non_smokers['expenses'], color='orange', label='Non-Smokers (Age <= 39)', alpha=0.7)

# Customizations
plt.title("Scatter Plot of Expenses by Age and Smoking Status", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Expenses", fontsize=12)
plt.legend(title="Group", fontsize=10)
plt.grid(True)

plt.show()
```

bargraph

```python
import matplotlib.pyplot as plt

# Calculate average expenses for each group
avg_expenses = {
    'Smokers (Age > 39)': above_39_smokers['expenses'].mean(),
    'Smokers (Age <= 39)': below_39_smokers['expenses'].mean(),
    'Non-Smokers (Age > 39)': above_39_non_smokers['expenses'].mean(),
    'Non-Smokers (Age <= 39)': below_39_non_smokers['expenses'].mean(),
}

# Prepare data for bar graph
groups = list(avg_expenses.keys())
avg_values = list(avg_expenses.values())

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(groups, avg_values, color=['red', 'blue', 'green', 'orange'], alpha=0.7)

# Customizations
plt.title("Average Expenses by Smoking Status and Age Group", fontsize=16)
plt.xlabel("Groups", fontsize=12)
plt.ylabel("Average Expenses", fontsize=12)
plt.xticks(rotation=15, fontsize=10)  # Rotate x-axis labels for clarity
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

plot graph age and bmi on x axis and expenses on y axis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Group by age and BMI, then calculate the mean expenses (charges) for each group
grouped_data = insurance_data.groupby(['age', 'bmi'])['expenses'].mean().reset_index()

# Create a pivot table for easier plotting, where 'age' and 'bmi' are the axes
pivot_data = grouped_data.pivot(index='age', columns='bmi', values='expenses')

# Plot the bar graph
pivot_data.plot(kind='bar', figsize=(12, 8))

# Customize the plot with titles and labels
plt.title("Average Insurance Charges by Age and BMI", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average Expenses ($)", fontsize=12)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
insurance_data = pd.read_csv("insurance.csv")

# Filter the data where charges are greater than 10,000
high_expenses_data = insurance_data[insurance_data['expenses'] > 10000]

# Separate the data for smokers and non-smokers
smokers = high_expenses_data[high_expenses_data['smoker'] == 'yes']
non_smokers = high_expenses_data[high_expenses_data['smoker'] == 'no']

# Count the number of smokers and non-smokers with expenses > 10,000
smokers_count = smokers.shape[0]
non_smokers_count = non_smokers.shape[0]

# Create a bar plot to show the comparison
categories = ['Smokers', 'Non-Smokers']
counts = [smokers_count, non_smokers_count]

plt.bar(categories, counts, color=['red', 'blue'])

# Annotate the bars with the count values
for i, count in enumerate(counts):
    plt.text(i, count + 1, str(count), ha='center', fontsize=12)

# Add title and labels
plt.title("Comparison of Smokers and Non-Smokers with Expenses > $10,000", fontsize=16)
plt.ylabel("Number of People", fontsize=12)

# Show the plot
plt.show()
```
